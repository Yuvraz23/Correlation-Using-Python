```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing dataset
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
#checking for null values
round(100*df.isnull().sum()/len(df),2)
```

```python
df[df.budget.isnull()]
```

```python
#Droping the null values
df1 = df.dropna()
```

```python
df1.info()
```

```python
df1.head()
```

```python
#now we will change the data type of the columns to its appropriate data type.
df1.votes = df1.votes.astype('int')
df1.budget = df1.budget.astype('int')
df1.gross = df1.gross.astype('int')
df1.runtime = df1.runtime.astype('int')
```

```python
df1.head()
```

```python
#now we will change the released column data type to date_time format
df1[['released','released_country']]=df1['released'].str.split('\(',n=1,expand=True)
```

```python
df1
```

```python
df1.released_country = df1.released_country.str.replace(')','')
```

```python
df1['year_corrected'] = df1.released.str[-5:]
```

```python
df1
```

```python
#Now we will drop year column as correction has been made

df1.drop(labels = 'year', axis = 1, inplace = True)
```

```python
df1[['company','budget']].groupby(by = 'company').sum()
```

```python
#we can see there are negative values in gross columns that needs to be treated
df1.sort_values(by='gross',ascending=True)
```

```python
#a negative gross is not possible anyways. So it needs to be changed into positive format.
df1['gross'] = df1['gross'].apply(abs)
```

```python
#now we will check again
df1.sort_values(by='gross',ascending=False)
```

```python
#since data is being treated properly now we will visualise the data
#ploting budget vs gross scatter plot
plt.style.use('ggplot')
plt.figure(figsize=[12,8])
sns.regplot(data = df1,x = 'budget', y= 'gross',scatter_kws = {'color':'red'}, line_kws = {'color':'blue'})
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize = [12,8])
sns.heatmap(df1.corr(),annot = True)
plt.show()
```

```python
#now we will se correlation between non-numeric colu,mns as well
numeric_df = df1.copy()
for cols in numeric_df.columns:
    if(numeric_df[cols].dtype == 'object'):
        numeric_df[cols] = numeric_df[cols].astype('category')
        numeric_df[cols] = numeric_df[cols].cat.codes

        
#printing new df
numeric_df        
```

```python
#since every column has been converted to numeric field
```

```python
pd.set_option('display.max_rows', None)
#now let see correlation of numeric df
numeric_df.corr().unstack()
```

```python
#lets visualize using heatmap
sns.set(font_scale = 3)
plt.figure(figsize = [60,30])
plt.title('Correlation Matrix')
sns.heatmap(numeric_df.corr(),annot = True, annot_kws = {'size':30}, cmap = 'Blues')
plt.show()
```

```python
# To understand correlation more properly
numeric_df.corr().unstack().sort_values(ascending = False)
```
